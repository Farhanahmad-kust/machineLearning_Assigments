```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("/content/drive/MyDrive/simple_linear_regression_salary_data.csv")
df
```

```python
plt.scatter(df.YearsExperience,df.Salary)
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
```

```python
lr = linear_model.LinearRegression()
lr.fit(df[['YearsExperience']],df.Salary)

```

```python
lr.predict([[2.5]])
```

```python
lr.score(df[['YearsExperience']],df.Salary)
```

```python
plt.plot(df.YearsExperience,lr.predict(df[['YearsExperience']]),color='red')
plt.scatter(df.YearsExperience,df.Salary)
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
```

```python
# now lets use that model in function to use it in real program where user can tell his year of experience and code will call the model and get the value and tell user what will be his salary based on his experience
while True:
  try:
    user_input = input("Enter your year of experience to get salary predictiom ")
    predict_value = lr.predict([[int(user_input)]])
    print(f"Your salary will be {predict_value}")

  except Exception as e:
    print(e)
    break




```
